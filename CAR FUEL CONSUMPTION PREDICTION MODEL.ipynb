<h4  style="color:violet;">  2.Data cleaning </h4>


```python
## treat missing and inappropriate values
missing_values = df.isnull().sum()
print('missing values\n',missing_values)
```

```python
# Drop unwanted columns
columns_to_drop = ['file','description','date_of_change']
df = df.drop(columns=columns_to_drop)
```

 The columns removed either contain non-informative or duplicate values (e.g., file could just be a file identifier,
description might overlap with other columns, and date_of_change likely doesn't contribute meaningfully to prediction).

```python
df.columns
```

```python
# finding duplicates
df.duplicated().sum()
```

```python
df = df.drop_duplicates()  # Removes duplicate rows

# Verify if duplicates have been removed
print(df.duplicated().sum()) 
```

```python
df.shape
```

```python
# Check the percentage of missing values in each column
missing_values = df.isnull().mean() * 100

# Filter columns with missing values
missing_columns = missing_values[missing_values > 0]

# Display columns with missing values and their percentages
print(missing_columns)
```

 This calculates the percentage of missing values in each column and
filters out the columns that have missing data. It identifies how many values are missing per feature and in what proportion.

```python
# Drop columns with more than 50% missing values
threshold = 50
columns_to_drop = missing_columns[missing_columns > threshold].index
df_cleaned = df.drop(columns=columns_to_drop)
print(df_cleaned)
```

```python
## check balance missing and inappropriate values
missing_values = df_cleaned.isnull().sum()
print('missing values\n',missing_values)
```

```python
# Handle missing values using mode imputation for all categorical columns
for column in df_cleaned.columns:
    df_cleaned[column] = df_cleaned[column].fillna(df_cleaned[column].mode().iloc[0])

# Check again to ensure missing values are handled
print(df_cleaned.isnull().sum())
```

```python
df_cleaned.shape
```

```python
df_cleaned.columns
```

```python
# Get column names
columns = df_cleaned.columns

# Analyze each column
for column in columns:
    print(f"--- {column} ---")
    print(df_cleaned[column].value_counts())
    print()
    
    # Plot bar chart
    plt.figure(figsize=(15, 9))
    sns.countplot(x=column, data=df_cleaned)
    plt.title(f"Distribution of {column}")
    plt.xticks(rotation=90)
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
# Label Encoding for all categorical columns
le = LabelEncoder()
for column in df_cleaned.columns:
    df_cleaned[column] = le.fit_transform(df_cleaned[column])
```

```python
# Box Plots (Before Outlier Removal)
for column in df_cleaned.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(df_cleaned[column])
    plt.title(f"Box Plot of {column} Before Outlier Removal")
    plt.show()
```

```python
# Correlation Matrix (Before Outlier Removal)
corr_matrix_before = df_cleaned.corr(method='spearman')
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix_before, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Before Outlier Removal")
plt.show()
```
